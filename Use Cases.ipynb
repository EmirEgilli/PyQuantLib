# PyQuantLib

## A library of some of the functions i most commonly use. Planning on frequent updates.

```python
#Import library
import PyQuantLib as PQL
```

### For Single Assets:

```python
PQL.MonteCarlo("BZ=F")
```

```python
PQL.VaR("BZ=F")
```

### For Stock Portfolios:

```python
PQL.StocksAnalyze("MA, IBM, XOM, KO, LMT", "2018-01-01")
```

### We can run an Efficient Frontier with 10000 allocations

```python
PQL.markowitz("MA, IBM, XOM, KO, LMT", "2018-01-01")
```

### Also with a Two-Fund Efficient Frontier to check for shorting opportunities and see if it improves the sharpe ratio

```python
PQL.TFEF("MA, IBM, XOM, KO, LMT")
```

### Additional verification with Capital Asset Pricing Model (CAPM)

```python
PQL.CAPM("MA, IBM, XOM, KO, LMT")
```
