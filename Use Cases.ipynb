```python
# Import Library
import PyQuantLib as PQL
```

## For Single Assets:

```python
PQL.MonteCarlo("BZ=F")
```

```python
PQL.VaR("BZ=F")
```

```python
PQL.garch("BZ=F")
```

## For Stock Portfolios:

```python
PQL.StocksAnalyze("MA, IBM, XOM, KO, LMT", start="2018-01-01")
```

### We can then run an Efficient Frontier with 10000 allocations:

```python
PQL.markowitz("MA, IBM, XOM, KO, LMT", "2018-01-01")
```

### Since some stocks are given zero weights, let's run a Two-Fund Efficient Frontier to check for shorting opportunities and see if it improves the Sharpe ratio

```python
PQL.TFEF("MA, IBM, XOM, KO, LMT", "2021-01-01")
```

### Additional verification with Capital Asset Pricing Model (CAPM) to see expected returns in shorter term

```python
PQL.CAPM("MA, IBM, XOM, KO, LMT", "2021-01-01")
```

### Finally, let's check our Sharpe ratio if we would use different weights than TFEF

```python
port_weights = {'MA': 0.04, 'IBM': 0.16, 'XOM': 0.22, 'KO': 0.18, 'LMT': 0.40}
PQL.PortSharpe(port_weights, start="2021-01-01", market="US")
```

### And also, check the GARCH for the portfolio with the given weights

```python
PQL.PortGarch(port_weights, start="2021-01-01")
```
